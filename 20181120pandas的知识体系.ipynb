# 函数

1. map()
    * 应用到Series的每个元素
2. apply()
    * 应用到df的行或列，或Series
    * 列中最大值-最小值
3. applymap()
    * 数据框的每个元素
4. agg()
    * 不改变数据框的结构

```python
df['color'].map(lambda x:str(x)+'s').head()# Series 的每个元素
```

```python
df['color'].apply(lambda x:str(x)+'s').head()
```

```python
df.groupby(['color','gross'])['title_year','actor_3_facebook_likes'].apply(lambda x:x.max()-x.min()).head()
```

```python
df.applymap(lambda x:str(x)+'k').head()# 应用到df的每个元素
```

```python
df.groupby(['color','gross'])[['director_name','language']].agg(lambda x:x).head()
```
