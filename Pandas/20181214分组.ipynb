## aggregate函数

```python
amzn.aggregate(np.sum)
```

```python
amzn.head()
```

```python
amzn.groupby(lambda Date:Date.split('-')[1]).aggregate(np.sum)
```

```python
amzn.groupby(lambda Date:Date.split('-')[0]).aggregate([np.mean,np.size])
```

```python
amzn.groupby(lambda Date:Date.split('-')[0]).agg([np.sum,np.mean])
```
