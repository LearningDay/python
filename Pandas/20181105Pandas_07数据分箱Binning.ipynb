```python
import numpy as np
import pandas as pd
```

```python
score_list = np.random.randint(50,100,size=20)
```

```python
score_list
```

```python
bins = [0,59,70,80,100]
```

# 分箱

```python
score_cut = pd.cut(score_list,bins)
score_cut
```

```python
pd.value_counts(score_cut)# pd.value_counts()
```

## DataFrame分箱

```python
df = pd.DataFrame()
```

```python
df['score_list'] = score_list
```

```python
df['name'] = [pd.util.testing.rands(3) for i in range(20)]
```

```python
df
```

```python
pd.cut(df['score_list'],bins)
```

```python
df['Categories'] = pd.cut(df['score_list'],bins)
```
