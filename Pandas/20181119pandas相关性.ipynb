```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

* 判断相关性
	* 协方差
		* 协方差>0，两者正相关
		*  协方差<0，两者负相关
		* 协方差=0，两者独立无关
	* Pearson相关系数
		* 取值范围[-1,1]
			* 0.8-1.0
				* 极强相关
			* 0.6-0.8
				* 强相关
			* 0.4-0.6
				* 中等程度相关
			* 0.2-0.4
				* 弱相关
			* 0.0-0.2
                * 极弱相关与无相关
                
## 适用于Series和DataFrame
* .cov()
	* 	计算协方差矩阵
* .corr()
	* 	计算相关系数矩阵
	* 	系数
		* 	Peries
		* 	Spearman
		* 	Kendall

```python
s1 = pd.Series([3.04,22.93,12.75,22.6,12.33],index=['2008','2009','2010','2011','2012'])
```

```python
s2 = pd.Series([8.18,18.38,9.13,7.82,6.69],index=['2008','2009','2010','2011','2012'])
```

```python
s1.corr(s2)
```

```python
plt.plot(s1.index,s1,s2.index,s2)
plt.legend(('s1','s2'))
```
