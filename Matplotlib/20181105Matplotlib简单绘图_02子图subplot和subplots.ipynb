```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.rcParams['font.sans-serif']=['SimHei']# 显示中文，修改默认字体
plt.rcParams['axes.unicode_minus']=False #解决‘-’显示为方框的问题
plt.style.use('ggplot')
```

```python
x = np.linspace(0,5)
y1 = np.sin(np.pi*x)
y2 = np.sin(np.pi*x*2)
```

```python
plt.figure(figsize=(12,5))
plt.plot(x,y1,label='sin(pi*x)')
plt.ylabel('y1轴')
plt.plot(x,y2,label='sin(pi*2x)')
plt.xlabel('x轴')
#plt.ylabel('y轴')
plt.title('demo')
plt.legend()
```

# subplot

```python
plt.figure(figsize=(12,7))
plt.subplot(221)
plt.plot(x,y1,'r',label='sin(pi*2)')
plt.xlabel('x轴')
plt.legend()

plt.subplot(222)
plt.plot(y2,y1,'b*',label='sin(pi*2x)')
plt.legend()

plt.subplot(212)
plt.plot(y1,y2,'p--',label='y1-y2')
plt.legend()
plt.title('y1和y2')
```

# subplots

```python
figure, ax = plt.subplots(2,2)
ax[0][0].plot(x,y1,'b')
ax[1][1].plot(x,y2,'r--')
ax[1][0].plot(y2,y1,'p-')
ax[0][1].plot(y1,y2,'*')
```
