```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
%matplotlib inline
```

```python
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
plt.plot(x)
plt.plot(y,'bo')
```

```python
plt.plot(x,y,'*')
```

```python
plt.plot(x,y,'--')
```

```python
plt.plot(x,y,'g--')
```

# demo

```python
plt.figure(figsize=(12,4))
plt.plot(x,y,'r--',label='aaa')
plt.plot(x*2,y,'b',label='bbb')
plt.xlabel('xlabel')
plt.ylabel('y')
plt.title('this is a demo')
plt.legend()
```
