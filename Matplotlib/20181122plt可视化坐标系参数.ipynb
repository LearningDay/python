```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(10,100,10)
```

```python
y = x**2
```

```python
y2 = x*2+1
```

```python
y1 = np.cos(2)
```

```python
plt.figure()
plt.plot(x,y2,'c:x',markersize=10)
plt.grid()
plt.axis([-20,110,-10,220])# 设置坐标轴范围
plt.savefig('demo1',dpi=300)
plt.show()
```

```python
plt.style.use('ggplot')# 全局（之后的所有图表） 更改图表 显示风格
plt.figure(figsize=(8,6))
plt.plot(x,y,linestyle='-.',color='m',marker='d',markersize=10,markerfacecolor='#008010')
plt.xlabel('x轴',fontproperties='SimHei',fontsize=20)
plt.ylabel('y轴',fontproperties='SimHei',fontsize=20)
plt.title(r'$y=x^2$这是一个标题',fontproperties='SimHei',fontsize=30)
plt.text(25,2000,r'$mu=100$')#再任意位置添加文本
plt.annotate(r'$y=cos(x)$样式',xy=(70,5000),xytext=(50,8000),arrowprops=dict(facecolor='m',width=2,shrink=0.1),
             fontproperties='SimHei',fontsize=15)# 带箭头的注释
plt.savefig('demo2',dpi=600)#保存图片
plt.show()

```
