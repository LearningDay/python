```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#plt.style.use('ggplot')# 全局（之后的所有图表） 更改图表 显示风格
```

```python
x = np.linspace(0,14,100)
```

```python
y1 = np.sin(x)
```

```python
y2 = np.sin(x/2+1)
```

```python
plt.figure(figsize=(12,6))
plt.plot(x,y1,'m:x',markersize=10,linewidth=3.0,markerfacecolor='b')
plt.grid()# 背景方框
plt.axis([-1,15,-1.2,1.2])# 设置坐标轴范围
plt.savefig('demo1',dpi=300)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(x,y2,linestyle='-.',linewidth=3.0,color='m',marker='d',markersize=10,markerfacecolor='#008010')
plt.xlabel('x轴',fontproperties='SimHei',fontsize=20)
plt.ylabel('y轴',fontproperties='SimHei',fontsize=20)
plt.title(r'$y_2=\sin(\frac{x}{2}+1)$这是一个标题',fontproperties='SimHei',fontsize=30)
plt.text(0,-0.5,r'$mu=100$')#在任意位置添加文本
plt.annotate(r'$y_2=sin(\frac{x}{2}+1)$',xy=(10.3,0),xytext=(10,-0.75),
             arrowprops=dict(facecolor='b',width=5,shrink=0.1,connectionstyle='arc3,rad=.2'),#设置箭头参数：颜色、宽度、箭头收敛度、弯曲度
             fontproperties='SimHei',fontsize=15)# 带箭头的注释
plt.savefig('demo2',dpi=600)#保存图片
plt.show()
```

# 注释

```python
x1=np.linspace(-4,4,100)
y3=2*x1+1
```

```python
plt.figure(figsize=(8,6))
x0 = 1
y0 = 2*x0+1
plt.scatter(x=x0,y=y0,s=50,color='b')
plt.plot([x0,x0],[y0,0],'k--',lw=3)
plt.plot(x1,y3)

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

plt.annotate(r'$2*x+1=%s$'%y0,xy=(x0,y0),xytext=(x0+3,y0-3),arrowprops=dict(facecolor='b',width=5,shrink=0.1,connectionstyle='arc3,rad=.2'),
             fontproperties='SimHei',fontsize=15)# 带箭头的注释

# xytext=(x0+3,y0-3) 注释位置
# xy=(x0,y0) 箭头位置
# arrowprops=dict 设置箭头参数
    # facecolor='m',设置箭头线条颜色
    # shrink=0.1, 设置箭头→ 尖的收敛度
    # connectionstyle='arc3,rad=.2' 弯曲
    
plt.text(-4,8,r'$This\ is\ the\ some\ text.\ \mu\ \sigma_i\ \alpha_t$',fontdict={'size':20,'color':'red'})
```

# 调整坐标
* [防遮挡](https://morvanzhou.github.io/tutorials/data-manipulation/plt/2-7-tick-visibility/)

```python
x = np.linspace(-3, 3, 50)
y = 0.1*x

plt.figure()
# 在 plt 2.0.2 或更高的版本中, 设置 zorder 给 plot 在 z 轴方向排序
plt.plot(x, y, linewidth=10, zorder=1)
plt.ylim(-2, 2)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.7))
plt.show()
```
