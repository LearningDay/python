	* 谱图

	* 绘制X-Y的相关性函数

	* 步阶图

	* 直方图
* plt.contour(X,Y,Z,N)
	* 等值图

	* 垂直图

```python
import matplotlib.pyplot as plt
import numpy as np
```

# 绘图参数

## 饼图
* plt.pie()

```python
labels = 'Frogs','Hogs','Dogs','Logs'# 定义饼图每块的标签
```

```python
type(labels)
```

```python
sizes = [15,30,45,10]# 饼图各块的尺寸
```

```python
explode = (0,0.1,0,0)# 突出显示哪一块，将第二块，再饼图中突出，突出比例0.1
```

```python
plt.pie(sizes,explode = explode,labels = labels,autopct='%1.1f%%',#autopct=,中线显示百分数的方式
       shadow = False,startangle=90)#shadows=False饼图不带阴影，startangle 饼图起始角度
plt.axis('equal')#让饼图xy方向尺寸相等
plt.show()
```

## 直方图
* plt.hist()

```python
np.random.seed(0)
mu, sigma = 100,20 #均值和标准差
```

```python
a = np.random.normal(mu,sigma,size=100)
```

```python
plt.hist(a,40,normed=1,# normed=1每个元素出现的比例， normed=0出现的概率
        histtype='stepfilled', #直方图的绘制类型
         facecolor='b',
         alpha = 0.75
        )
plt.title('直方图',fontproperties='SimHei',fontsize=50,color='m')
plt.show()
```

## 极坐标图
* plt.polar()
* 不常用

```python
N = 10
theta = np.linspace(0.0,2*np.pi,N,endpoint=False)
```

```python
radii = 10 * np.random.rand(N)
```

```python
width = np.pi/4 * np.random.rand(N)
```

```python
ax = plt.subplot(111,projection='polar') # projection='polar' 极坐标
bars = ax.bar(theta,radii, width = width, bottom=0.0)

for r, bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.viridis(r/10.))
    bar.set_alpha(0.5)

plt.show()
```
