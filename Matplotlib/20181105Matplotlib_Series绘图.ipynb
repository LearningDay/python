```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.rcParams['font.sans-serif']=['SimHei']# 显示中文，修改默认字体
plt.rcParams['axes.unicode_minus']=False #解决‘-’显示为方框的问题
```

```python
s1 = pd.Series(np.random.randn(100)).cumsum()
s2 = pd.Series(np.random.randn(100)).cumsum()
```

```python
plt.plot(s1)
```

# Series绘图

```python
s1.plot()
```

```python
s1.plot(kind='line',grid=True,label='s1',title='Series',style='--')
plt.legend()
```

```python
s1.plot(label='S1')
s2.plot(label='S2')
plt.legend()
```

# 子图
* 方法一

```python
fig, ax= plt.subplots(2,1)
ax[0].plot(s1)
ax[1].plot(s2)
```

* 方法二
* 推荐

```python
fig, ax = plt.subplots(2,1)
s1[:10].plot(ax=ax[0],label='S1',kind='bar')
s2.plot(ax=ax[1],label='S2',figsize=(12,6))
```
