```python
df = pd.DataFrame(np.random.randint(10,40,40).reshape(10,4),columns=list('ABCD'))
```

```python
df
```

```python
df.plot(kind='bar')
```

```python
df.plot(kind='box')
```

## 对行绘图

```python
df.iloc[5].plot()
```

```python
for i in df.index:
    df.iloc[i].plot(label=str(i))
plt.legend()
plt.show()
```

# 转置
* 实现行画图

```python
df.T.plot()
```

## 列绘图
* 默认

```python
df['A'].plot()
```

```python
for col in df.columns:
    df.loc[:,col].plot(label=str(col))
plt.legend()
plt.show()
```

```python
df.plot()
```

## 散点图矩阵

```python
pd.plotting.scatter_matrix(df,diagonal='kde',figsize=(12,12))
```

```python
pd.plotting.scatter_matrix(df[['A','B','C']],diagonal='hist',figsize=(12,12))
```
