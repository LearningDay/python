# matplotlib 绘图

```python
plt.figure(1,figsize=(6,6))
plt.pie(df.groupby('city').avg.count(),labels=df.groupby('city').avg.count().index)
plt.title('这是一个饼图')
plt.show() # 不显示内存地址，只显示图形
```

```python
plt.figure(1,figsize=(10,4))
plt.plot(np.random.randint(20,100,50))
plt.xticks([-10,0,10,60])
plt.yticks(np.arange(-10,110,10))
plt.title('这是一条折线图')
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.show()
```

```python
plt.figure(1,figsize=(10,4))
plt.plot(np.random.randint(20,100,50),label='no1',color='g')
plt.plot(np.random.randint(20,100,50),label='no2',color='r')
plt.xticks([-10,0,10,60])
plt.yticks(np.arange(-10,110,10))
plt.title('这是两条折线图')
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.legend()#显示图例
plt.show()
```

```python
data_1 = df.groupby(['city','companyId']).aggregate(['mean','count']).avg.reset_index()
data_1.head()
```

* 不同城市下，每个公司招聘人数和平均薪资

```python
for city,grouped in data_1.groupby('city'):
    x = grouped['mean']
    y = grouped['count']
    plt.figure(1,figsize=(6,6))
    plt.scatter(x,y,label=city,alpha=0.5)
    plt.legend()
plt.xlabel('平均薪资')
plt.ylabel('招聘人数')
plt.show()
```
