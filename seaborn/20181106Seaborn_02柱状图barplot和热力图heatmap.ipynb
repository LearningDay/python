```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('flights')
```

```python
df.shape
```

```python
df.head()
```

## 看各个整年的数据

```python
df1 = df.pivot_table(index='month',columns='year',values='passengers',aggfunc='mean')
```

```python
df1
```

```python
sns.heatmap(df1)
```

```python
df1.plot()
```

## 显示数字
* 整数

```python
sns.heatmap(df1,annot=True,fmt='d')
```
