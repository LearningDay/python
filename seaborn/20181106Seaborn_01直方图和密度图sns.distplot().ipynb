```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
s1 = pd.Series(np.random.randn(1000))
```

```python
plt.hist(s1)
```

```python
s1.plot(kind='kde')
```

# Seaborn

```python
sns.distplot(s1,hist=True,kde=True,rug=True,bins=20,axlabel='x')
```

## sns.kdeplot()

```python
sns.kdeplot(s1,shade=True,vertical=True,color='r')
```

## sns.rugplot()

```python
sns.rugplot(s1)
```
